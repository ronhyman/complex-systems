```python
import numpy as np
import matplotlib.pyplot as plt


# (a) Competing Dynamics Model
# x  = ( 3  x  2 y ) x
# y  = ( 2  y  x ) y

# params
delta_t = 0.1
time_steps = 1000
 = 2

# Function for model A
def model_a(x, y):
    dx = (3 - x - 2*y) * x
    dy = (2 - y - x) * y
    return dx, dy

# Simulate from a starting point
def simulate(start_x, start_y, model):
    x, y = start_x, start_y
    xs, ys = [x], [y]
    for _ in range(time_steps):
        dx, dy = model(x, y)
        x += dx * delta_t
        y += dy * delta_t
        xs.append(x)
        ys.append(y)
    return xs, ys

def plot_phase_diagram(model, start_x, start_y):
    xs, ys = simulate(start_x, start_y, model)
    plt.plot(xs, ys, '-')
    plt.show()


# Plotting A
fig, ax = plt.subplots()
for _ in range(100):
    start_x, start_y = np.random.uniform(0, 3), np.random.uniform(0, 2)
    xs, ys = simulate(start_x, start_y, model_a)
    ax.plot(xs, ys, '-')

ax.set_xlim([0, 3])
ax.set_ylim([0, 2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Phase Diagram for model_a')
ax.set_prop_cycle(None)
plt.show()
```

```python
# B) Predator-Prey Model

# Function for model B
def model_b(x, y):
    dx = (1 - y) * x
    dy = *(x-1)*y* #note that theta is > 0
    return dx, dy

# Plotting B
fig, ax = plt.subplots()
for _ in range(100):
    start_x, start_y = np.random.uniform(0, 1), np.random.uniform(0, 1)
    xs, ys = simulate(start_x, start_y, model_b)
    ax.plot(xs, ys, '-')
ax.set_xlim([0, 2])
ax.set_ylim([0, 3])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Phase Diagram for model_b')
plt.show()
```

```python
# Find behaviour of model c at fixed pt (ii)

import numpy as np
import matplotlib.pyplot as plt

# Define theta range
theta_range = np.linspace(0, 1, 100)

# Calculate tau and Delta for each theta
tau = -1 + 2 * theta_range * (theta_range / (0.1 + theta_range**2)) - 0.1 - theta_range**2
Delta = (-1 + 2 * theta_range * (theta_range / (0.1 + theta_range**2))) * (-0.1 - theta_range**2) - (0.1 + theta_range**2) * (-2 * theta_range * (theta_range / (0.1 + theta_range**2)))

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(theta_range, tau, label=' (Trace)')
plt.plot(theta_range, Delta, label=' (Determinant)')
plt.xlabel('')
plt.ylabel('Value')
plt.title(' (Trace) and  (Determinant) vs. ')
plt.legend()
plt.grid(True)
plt.show()

# plot on a - plot.
plt.figure(figsize=(8, 4))
plt.plot(Delta, tau)
plt.xlabel(' (Determinant)')
plt.ylabel(' (Trace)')
plt.title(' (Trace) vs.  (Determinant)')
plt.grid(True)
plt.show()

```
